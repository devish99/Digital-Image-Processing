# Scale Invariant Feature Transform

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(18,10))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

image = cv2.imread('C:/Users/psvis/OneDrive/Pictures/Saved pictures/Eagle.jpg')
src = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(src)

sift = cv2.xfeatures2d.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(src, None)

kp_img = cv2.drawKeypoints(src, keypoints, None)
kp_img_2 = cv2.drawKeypoints(src, keypoints, None, color=(0, 255, 0), flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

ax1.title.set_text('Original')
ax1.imshow(src)

ax2.title.set_text('SIFT Output-1')
ax2.imshow(kp_img)

ax3.title.set_text('SIFT Output-2')
ax3.imshow(kp_img_2)

plt.show()
```
