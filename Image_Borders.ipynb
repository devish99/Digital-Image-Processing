# Experiment 2: Adding Borders and Padding to Images
## Creating Borders for Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('C:/Users/psvis/OneDrive/Pictures/Windows XP HD Wallpaper Pack 4.0/Windows XP/Tulips.jpg')


const_border = [255, 0, 180]

image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
replicate = cv2.copyMakeBorder(image,100,100,100,100,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(image,100,100,100,100,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(image,100,100,100,100,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(image,60,60,60,60,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(image,80,80,80,80,cv2.BORDER_CONSTANT,value=const_border)

```

## Displaying Borders

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(231)
ax2 = fig.add_subplot(232)
ax3 = fig.add_subplot(233)
ax4 = fig.add_subplot(234)
ax5 = fig.add_subplot(235)
ax6 = fig.add_subplot(236)

ax1.imshow(image,'gray')
ax1.title.set_text('ORIGINAL')

ax2.imshow(replicate,'gray')
ax2.title.set_text('REPLICATE')

ax3.imshow(reflect,'gray')
ax3.title.set_text('REFLECT')

ax4.imshow(reflect101,'gray')
ax4.title.set_text('REFLECT_101')

ax5.imshow(wrap,'gray')
ax5.title.set_text('WRAP')

ax6.imshow(constant,'gray')
ax6.title.set_text('CONSTANT')

plt.subplots_adjust(hspace=0)
plt.show()
```

## Region Of Interest

```python
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)


#Displaying image
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
ax1.title.set_text('Original Image')
ax1.imshow(image)


# Accesssing a Region of Interest
petals = image[180:480, 180:420]
image =cv2.cvtColor(petals, cv2.COLOR_BGR2RGB)
ax2.title.set_text('ROI: Petals')
ax2.imshow(petals)

plt.show()
```

## RGB Channel Splitting

```python
# Splitting Image Channels
b, g, r = cv2.split(image)
rgb_image = cv2.merge((r, g, b))
```

```python
# Displaying Image Channels
fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)

ax1.imshow(image)
ax1.title.set_text('BGR Image')
ax2.imshow(rgb_image)
ax2.title.set_text('RGB Image')
plt.show()
```

## Addition And Subtraction Of Images

```python
fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)

ax1.title.set_text('Addition')
ax1.imshow(rgb_image+image)

ax2.title.set_text('Subtraction')
ax2.imshow(rgb_image-image)

plt.show()
```
