```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('C:/Users/psvis/OneDrive/Pictures/Windows XP HD Wallpaper Pack 4.0/Windows XP/Tulips.jpg')


const_border = [255, 0, 180]

image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
replicate = cv2.copyMakeBorder(image,80,80,80,80,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(image,80,80,80,80,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(image,80,80,80,80,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(image,40,40,40,40,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(image,40,40,40,40,cv2.BORDER_CONSTANT,value=const_border)

plt.imshow(image,'gray')
plt.title('ORIGINAL')
plt.show()
```
