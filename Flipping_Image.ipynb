```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('C:/Users/psvis/OneDrive/Pictures/Reddit Wallpapers/Calvin_Hobbes.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
# Mirroring the image
mirrored = cv2.flip(image, 1)
plt.imshow(mirrored)
plt.show()
```

```python
# Flipping the image
flipped = cv2.flip(image, 0)
plt.imshow(flipped)
plt.show()
```

```python
flipped = cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB)
mirrored = cv2.cvtColor(mirrored, cv2.COLOR_BGR2RGB)
# Writing images into a folder
cv2.imwrite('C:/Users/psvis/OneDrive/Documents/Sem 7/Digital Image Processing Lab/Flipping_Image_Lab_2/flipped.jpg', flipped)
cv2.imwrite('C:/Users/psvis/OneDrive/Documents/Sem 7/Digital Image Processing Lab/Flipping_Image_Lab_2/mirrored.jpg', mirrored)
```
