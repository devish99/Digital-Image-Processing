## Erosion, Dilation and Gradient

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

image = cv2.imread("C:/Users/psvis/OneDrive/Pictures/Reddit Wallpapers/Grid.jpg")
src = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

k = cv2.getStructuringElement(cv2.MORPH_CROSS,(5, 5))
erosion = cv2.erode(src, k, iterations = 5)
dilation = cv2.dilate(src, k, iterations = 5)
gradient = cv2.morphologyEx(src, cv2.MORPH_GRADIENT, k)

ax1.title.set_text("Original")
ax1.imshow(src)
ax2.title.set_text("Erosion")    
ax2.imshow(erosion)
ax3.title.set_text("Dilation")
ax3.imshow(dilation)
ax4.title.set_text("Gradient")
ax4.imshow(gradient)

plt.show()
```

## Opening and Closing

```python
fig = plt.figure(figsize=(14,10))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

k = cv2.getStructuringElement(cv2.MORPH_CROSS,(8, 8))
opening = cv2.morphologyEx(src, cv2.MORPH_OPEN, k)

k = cv2.getStructuringElement(cv2.MORPH_CROSS,(8, 8))
closing = cv2.morphologyEx(src, cv2.MORPH_CLOSE, k)

ax1.title.set_text("Original")
ax1.imshow(src)
ax2.title.set_text("Opening")    
ax2.imshow(opening)
ax3.title.set_text("Closing")
ax3.imshow(closing)

plt.show()
```
