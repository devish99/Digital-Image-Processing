# Experiment 2: Flipping, Mirroring, Color Spaces, Brightness and Contrast
## Flipping and Mirroring

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(231)
ax2 = fig.add_subplot(232)
ax3 = fig.add_subplot(233)


# Reading the Image
image = cv2.imread('C:/Users/psvis/OneDrive/Pictures/Reddit Wallpapers/Calvin_Hobbes.jpg')
src = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
ax1.title.set_text('ORIGINAL')
ax1.imshow(src)

# Flipping the image
flipped = cv2.flip(src, 0)
ax2.title.set_text('FLIPPED')
ax2.imshow(flipped)

# Mirroring the image
mirrored = cv2.flip(src, 1)
ax3.title.set_text('MIRRORED')
ax3.imshow(mirrored)

plt.show()
```

## Changing Color Spaces

```python
image = cv2.imread('C:/Users/psvis/OneDrive/Pictures/Reddit Wallpapers/Calvin_Hobbes.jpg')
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(231)
ax2 = fig.add_subplot(232)

# BGR to RGB
src = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
ax1.title.set_text('BGR to RGB')
ax1.imshow(src)

# BGR to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
ax2.title.set_text('BGR to HSV')
ax2.imshow(hsv)

plt.show()
```

## Contrast and Brightness

```python
src = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
array_alpha = np.array([2.0])
array_beta = np.array([100.0])

# add a beta value to every pixel 
src1=cv2.add(src, array_beta)                    

# multiply every pixel value by alpha
src2=cv2.multiply(src, array_alpha)
```

```python
fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

ax1.title.set_text('Original')
ax1.imshow(src)

ax2.title.set_text('Brightness Change')
ax2.imshow(src1)

ax3.title.set_text('Contrast Change')
ax3.imshow(src2)

plt.show()
```
